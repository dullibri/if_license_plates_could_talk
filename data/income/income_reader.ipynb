## Arbeitnehmerentgelt auf Kreisebene

```python
df = pd.read_excel("vgr_r2b2_bs2020_2.xlsx", sheet_name = "1.1", skiprows = 4)
```

```python
df.head()
```

```python
df.rename(columns = {"Regional-schlüssel": "kreis_key", 2019 : "income"}, inplace = True)
df = df[["kreis_key", "income"]].dropna()
df.kreis_key = df.kreis_key.astype(int)
df.kreis_key.replace({2 : 2000, 11 : 11000}, inplace = True) # Fix Regional codes of Hamburg (2) and Berlin (11)
```

```python
df.to_csv("income.csv")
```
