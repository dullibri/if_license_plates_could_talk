```python
import pandas as pd
```

## Lade Daten

```python
df_raw = pd.read_csv("82411-01-03-4.csv", encoding="ISO-8859-1", skiprows = 6, delimiter = ";")
```

```python
# clean data

df_raw = df_raw.rename(columns = {"Unnamed: 0" : "year", "Unnamed: 1" : "kreis_key", "Unnamed: 2" : "kreis_name", "Tsd. EUR" : "income", "EUR" : "income_pp"})
df_raw = df_raw.dropna(subset = ["kreis_key"])
df_raw = df_raw[df_raw.kreis_key != "DG"]
df_raw.kreis_key.replace({"02": "02000", "11" : "11000"}, inplace=True) # Berlin, Hamburg

df_raw.head()
```

```python
# Pivot Data
df_piv = df_raw.pivot(index = "kreis_key", columns = "year", values = ["income_pp"])

df_piv.columns = df_piv.columns.droplevel(0)

df_piv = df_piv.reset_index()

df_piv = df_piv.rename(columns =  {str(i) : f"income_pp_{i}" for i in df_raw.year.unique()})
```

```python
df_piv.head()
```

```python
# drop rows that are not on "Kreis"-level

df = df_piv[[len(i) == 5 for i in df_piv.kreis_key]]
```

```python
df.head()
```

```python
# save to csv

df.to_csv("income.csv")
```
